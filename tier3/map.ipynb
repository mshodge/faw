```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('teams.csv')
df

# Make an empty map
m = folium.Map(location=[52.458971, -3.381157], tiles="CartoDB positron", zoom_start=8)
 
    
colordict = {'South East': 'lightblue', 'South West': 'lightgreen', 'North East': 'orange', 'North West': 'red'}
    
style = {'fillColor': '#fff', 'color': '#000', 'opacity': '0.1', 'fill_opacity': '0.1'}

folium.TopoJson(
    open('topo_nawc.json'),
    object_path='objects.nawc',
    style_function=lambda x: style,
    tooltip=folium.GeoJsonTooltip(fields=['NAWC13NM'],
                                             aliases=['Area']),
).add_to(m)

# I can add marker one by one on the map
for lat, lon, team, league in zip(df['lat'], df['lon'], df['Team'], df['League']):
   folium.CircleMarker(
      location=[lat, lon],
      radius=5,
       stroke=True,
       popup=f"{team} ({league})",
       color='black',
       opacity=1,
       weight=1,
       fill=True,
      fill_color=colordict[league],
       fill_opacity=1
   ).add_to(m)



m
```

```python
m.save(outfile= "index.html")
```
